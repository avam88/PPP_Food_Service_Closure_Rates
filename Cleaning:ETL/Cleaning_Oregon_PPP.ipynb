```python
# Add the Pandas dependency.
import pandas as pd
```

```python
# Files to load
under_loan_data_df = pd.read_csv("data/public_up_to_150k_no_dupes.csv")
over_loan_data_df = pd.read_csv("data/public_150k_plus_no_dupes.csv")
all_or_ppp = under_loan_data_df.append(over_loan_data_df, ignore_index=True)
all_or_ppp
```

```python
# dropping and reorganizing columns in dataframe
new_column_order = ["BorrowerName", "BorrowerAddress", "BorrowerCity", "BorrowerState", "BorrowerZip", "JobsReported", "NAICSCode", "CurrentApprovalAmount", "ForgivenessAmount", "DateApproved"]
loan_data_df = all_or_ppp[new_column_order]
print("All PPP Loan Data")
loan_data_df.head()
```

```python
loan_data_df = loan_data_df.fillna(0)
loan_data_df
```

```python
# filtering loans under $150,000 for just OR

oregon_data_df = loan_data_df[loan_data_df["BorrowerState"] == "OR"]
print("All PPP Loan Data for Oregon")
oregon_data_df
```

```python
oregon_df = oregon_data_df.rename(columns={"BorrowerName":"Business Name","BorrowerAddress":"Street Address","BorrowerCity":"City","BorrowerState":"State", "BorrowerZip":"Zip", "CurrentApprovalAmount":"Loan Amount", "ForgivenessAmount": "Foregiveness Amount", "DateApproved": "Approval Date", "JobsReported": "Jobs", "NAICSCode":"NAICS Code"})
new_column_order = ["Business Name", "Street Address", "City", "State", "Zip", "Loan Amount", "Foregiveness Amount", "Approval Date", "Jobs", "NAICS Code"]
oregon_df = oregon_df[new_column_order]
oregon_df["Zip"] = oregon_df["Zip"].str.extract(r'(\d{5})')
oregon_df = oregon_df.fillna(0)
oregon_df
```

```python
oregon_df["Foregiveness Rate"] = oregon_df["Foregiveness Amount"]/oregon_df["Loan Amount"]*100
oregon_df["Foregiveness Rate"] = oregon_df["Foregiveness Rate"].map("{:,.2f}".format)
oregon_df
```

```python
oregon_df = oregon_df.astype({'NAICS Code': 'str'})
oregon_df["NAICS Code"] = oregon_df["NAICS Code"].str.extract(r'(\d{2})')
oregon_df
```

```python
oregon_df = oregon_df.dropna()
oregon_df
```

```python
NAICS_df = pd.read_csv("data/NAICS Codes.csv")
NAICS_df
```

```python
key_list = list(oregon_df["NAICS Code"])
key_list
```

```python
NAICS_df = NAICS_df.astype({'Code': 'str'})
dict_lookup = dict(zip(NAICS_df['Code'], NAICS_df['Industry']))
dict_lookup
```

```python
oregon_df["NAICS Code"] = [dict_lookup[item] for item in key_list]
oregon_df
```

```python
output_data_file = "all_oregon_ppp_data.csv"
# Export the City_Data into a CSV.
oregon_df.to_csv(output_data_file, index=False)
```

```python
# filtering Oregon loan applications for food establishments only with NAICS code for food establishments
oregon_restaurant_df = oregon_data_df[(oregon_data_df["NAICSCode"] == 722320)|(oregon_data_df["NAICSCode"] == 722330) |(oregon_data_df["NAICSCode"] == 722410)| (oregon_data_df["NAICSCode"] == 722511)| (oregon_data_df["NAICSCode"] == 722513)| (oregon_data_df["NAICSCode"] == 722514)| (oregon_data_df["NAICSCode"] == 722515)]
print("All PPP Loan Data for Oregon Food Establishments")
oregon_restaurant_df
```

```python
# establishing filtered dataset size is still adequate to perform analysis on
oregon_restaurant_count = oregon_restaurant_df["BorrowerName"].count()
print("Count of all Oregon Food Establishments that received PPP Loans")
oregon_restaurant_count
```

```python
# creating a dataframe of all Oregon business that received PPP funding that are NOT food establishments
oregon_data_df = oregon_data_df[(oregon_data_df["NAICSCode"] != 722320)&(oregon_data_df["NAICSCode"] != 722330) &(oregon_data_df["NAICSCode"] != 722410)& (oregon_data_df["NAICSCode"] != 722511)& (oregon_data_df["NAICSCode"] != 722513)& (oregon_data_df["NAICSCode"] != 722514)&(oregon_data_df["NAICSCode"] != 722515)]
print("Oregon PPP Loan Data for All Business Except Food Establishments")
oregon_data_df.head()
```

```python
# establishing count of oregon business that are not food establihsments that received funding
oregon_loan_count = oregon_data_df["BorrowerName"].count()
print("Count of all Oregon non food establishments business that recieved PPP funding")
oregon_loan_count
```

```python
print("The mean Loan for Oregon Restaurants")
mean_resto = oregon_restaurant_df["CurrentApprovalAmount"].mean()
mean_resto
```

```python
print("The mean Loan for Oregon Business (Excluding Restaurants)")
mean = oregon_data_df["CurrentApprovalAmount"].mean()
mean
```

```python
# establishing foregiveness ratio for oregon restaurants loans
oregon_restaurant_foregiveness = oregon_restaurant_df["ForgivenessAmount"]/oregon_restaurant_df["CurrentApprovalAmount"]*100
oregon_restaurant_foregiveness.head()
```

```python
# establishing foregiveness ratio for oregon business loans excluding restaurants
oregon_foregiveness = oregon_data_df["ForgivenessAmount"]/oregon_data_df["CurrentApprovalAmount"]*100
oregon_foregiveness.head()
```

```python
print("The Mean Foregiveness Ratio for Oregon Restaurants")
rest_foregive_ratio = oregon_restaurant_foregiveness.mean()
rest_foregive_ratio
```

```python
print("The Mean Foregiveness Ratio for Oregon Businesses (Excluding Restaurants)")
foregive_ratio = oregon_foregiveness.mean()
foregive_ratio
```

```python
oregon_loan_sum = oregon_data_df["CurrentApprovalAmount"].sum()
oregon_loan_sum
```

```python
oregon_restaurant_sum = oregon_restaurant_df["CurrentApprovalAmount"].sum()
oregon_restaurant_sum
```

```python
loan_sum_ratio = (oregon_restaurant_sum/(oregon_restaurant_sum + oregon_loan_sum))*100
loan_sum_ratio
```

```python
business_count_ratio = (oregon_restaurant_count/(oregon_loan_count + oregon_restaurant_count))*100
business_count_ratio
```

```python
# create new dataframe where each row is oregon biz vs oregon resto and each column is a stat
state_ppp_summary_df = pd.DataFrame(
          [{"Loan Dispersal Count: Non Food Establishments": oregon_loan_count,
            "Loan Dsipersal Count: Food Establishments": oregon_restaurant_count,
            "Average Loan Value: Non Food Establishments": mean,
            "Average Loan Value: Food Establishments": mean_resto,
            "Foregiveness Ratio: Non Food Establishments": foregive_ratio,
            "Foregiveness Ratio: Food Establishments": rest_foregive_ratio,
            "Total Loan Dispersement: Non Food Establishments": oregon_loan_sum,
            "Total Loan Dispersement: Food Establishments": oregon_restaurant_sum,
            "Restaurant to Total: Business Count Ratio": business_count_ratio,
            "Restaurant to Total: Loan Amount Ratio": loan_sum_ratio
           }])
state_ppp_summary_df
```

```python
state_ppp_summary_df["Average Loan Value: Non Food Establishments"] = state_ppp_summary_df["Average Loan Value: Non Food Establishments"].map("${:,.2f}".format)
state_ppp_summary_df["Average Loan Value: Food Establishments"] = state_ppp_summary_df["Average Loan Value: Food Establishments"].map("${:,.2f}".format)
state_ppp_summary_df["Foregiveness Ratio: Non Food Establishments"] = state_ppp_summary_df["Foregiveness Ratio: Non Food Establishments"].map("{:.1f}".format)
state_ppp_summary_df["Foregiveness Ratio: Food Establishments"] = state_ppp_summary_df["Foregiveness Ratio: Food Establishments"].map("{:.1f}".format)
state_ppp_summary_df["Total Loan Dispersement: Non Food Establishments"] = state_ppp_summary_df["Total Loan Dispersement: Non Food Establishments"].map("${:,.2f}".format)
state_ppp_summary_df["Total Loan Dispersement: Food Establishments"] = state_ppp_summary_df["Total Loan Dispersement: Food Establishments"].map("${:,.2f}".format)
state_ppp_summary_df["Restaurant to Total: Business Count Ratio"] = state_ppp_summary_df["Restaurant to Total: Business Count Ratio"].map("{:.1f}%".format)
state_ppp_summary_df["Restaurant to Total: Loan Amount Ratio"] = state_ppp_summary_df["Restaurant to Total: Loan Amount Ratio"].map("{:.1f}%".format)
state_ppp_summary_df

# district_summary_df["Total Budget"] = district_summary_df["Total Budget"].map("${:,.2f}".format)
```

```python
# renaming columns to prepare for yelp_API dataframe merge
oregon_restaurant_df = oregon_restaurant_df.rename(columns={"BorrowerName":"Business Name","BorrowerAddress":"Street Address","BorrowerCity":"City","BorrowerState":"State", "BorrowerZip":"Zip", "CurrentApprovalAmount":"Loan Amount", "ForgivenessAmount": "Foregiveness Amount", "DateApproved": "Approval Date", "JobsReported": "Jobs"})
oregon_restaurant_df.columns
```

```python
# preparing dataframe to concatenate street address to match formatting with google_API dataframe address column
new_column_order = ["Business Name", "Street Address", "City", "State", "Zip", "Loan Amount", "Foregiveness Amount", "Approval Date", "Jobs"]
oregon_restaurant_df = oregon_restaurant_df[new_column_order]

oregon_restaurant_df.head()
```

```python
# regular expression on zip code to establish 5 digit standard
oregon_restaurant_df["Zip"] = oregon_restaurant_df["Zip"].str.extract(r'(\d{5})')
```

```python
# concatenate address columns to reflect idential formatting of yelp/OR biz dataframe 
Address = []
for index in range(len(oregon_restaurant_df)):
    address = f'{oregon_restaurant_df.iloc[index, 1]} , {oregon_restaurant_df.iloc[index, 2]}, {oregon_restaurant_df.iloc[index, 3]} {oregon_restaurant_df.iloc[index, 4]} ,  United States'
    Address.append(address)
   
len(Address)
```

```python
# establishing new address column in dataframe and reformatting columns
oregon_restaurant_df['Address'] = Address
new_column_order = ["Business Name", "Address", "Loan Amount", "Foregiveness Amount", "Approval Date", "Jobs"]
oregon_restaurant_df = oregon_restaurant_df[new_column_order]

oregon_restaurant_df = oregon_restaurant_df.fillna(0)
oregon_restaurant_df.head(20)

                                     
```

```python
oregon_restaurant_df["Address"] = oregon_restaurant_df["Address"].str.upper()
oregon_restaurant_df
```

```python
all_or_food_ppp = oregon_restaurant_df.groupby(["Business Name", "Address"]).agg({"Loan Amount": ['sum'], "Foregiveness Amount": ['sum'], "Jobs":['sum'], "Approval Date":['min']}).reset_index()
all_or_food_ppp
```

```python
all_or_food_ppp["Foregiveness Rate"] = all_or_food_ppp["Foregiveness Amount"]/all_or_food_ppp["Loan Amount"]*100
all_or_food_ppp
```

```python
all_or_food_ppp["Foregiveness Rate"] = all_or_food_ppp["Foregiveness Rate"].map("{:,.4f}".format)
all_or_food_ppp
```
