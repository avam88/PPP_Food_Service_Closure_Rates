```python
import pandas as pd
import numpy as np
```

```python
all_or_food_ppp = pd.read_csv("data/all_oregon_restaurant_ppp_data.csv")
or_yelp_food_df = pd.read_csv("data/final_oregon_restaurant_yelp_data_combined.csv")
oregon_mail_df = pd.read_csv("data/final_oregon_allbiz_mailling_data.csv")
all_or_food_ppp
```

```python
# dataframe for machine learning. It is only the ppp data, using jobs/funding amount and ratio, we can determine if it will be open or closed
ppp_yelp_left_ppp = all_or_food_ppp.merge(or_yelp_food_df,  how='left', on=['Address'])
ppp_yelp_left_ppp["New Business Status"] = ""
ppp_yelp_left_ppp['Business Status'] = ppp_yelp_left_ppp['Business Status'].map({False: 'False'})
for index in range(len(ppp_yelp_left_ppp)):
    if ppp_yelp_left_ppp["Business Status"][index] == "False":
        ppp_yelp_left_ppp["New Business Status"][index] = "Open"
    else:
        ppp_yelp_left_ppp["New Business Status"][index] = "Closed"
ppp_yelp_left_ppp
```

```python
ppp_yelp_left_ppp = ppp_yelp_left_ppp.drop([0], axis=0)
ppp_yelp_left_ppp
```

```python
new_column_order = ["Address", 
                    "Business Name_x", 
                    "Loan Amount", 
                    "Foregiveness Amount", 
                    "Foregiveness Rate", 
                    "Approval Date", 
                    "New Business Status", 
                    "Jobs"]
ppp_yelp_left_ppp = ppp_yelp_left_ppp[new_column_order]
ppp_yelp_left_ppp = ppp_yelp_left_ppp.rename(columns={"Business Name_x":"Business Name (PPP)",
                                                            "BorrowerZip":"Zip", 
                                                            "New Business Status":"Business Status"})
ppp_yelp_left_ppp
```

```python
address = ppp_yelp_left_ppp["Address"].str.split(", ", expand = True)
address
```

```python
zipcode = address[2].str.split(" ", expand = True)
zipcode
```

```python
# making separate first name column from new data frame
ppp_yelp_left_ppp["Zipcode"]= zipcode[1]
ppp_yelp_left_ppp["City"] = address[1]
ppp_yelp_left_ppp.iloc[2, 8] = '97013'
# making separate last name column from new data frame
ppp_yelp_left_ppp
```

```python
ppp_yelp_allbiz_df = ppp_yelp_left_ppp.merge(oregon_mail_df,  how='left', on=['Address'])
ppp_yelp_allbiz_df
```

```python
ppp_yelp_allbiz_df["Business Name"].fillna('Closed', inplace = True)
ppp_yelp_allbiz_df
```

```python
for index in range(len(ppp_yelp_allbiz_df)):
    if ppp_yelp_allbiz_df["Business Name"][index] != "Closed":
        ppp_yelp_allbiz_df["Business Status"][index] = "Open"

ppp_yelp_allbiz_df
```

```python
new_column_order = ["Address", 
                    "Business Name (PPP)",
                    "Loan Amount", 
                    "Foregiveness Amount", 
                    "Foregiveness Rate", 
                    "Approval Date", 
                    "Business Status",
                    "Jobs",
                    "Zipcode",
                    "City"]
                   
ppp_yelp_allbiz_df = ppp_yelp_allbiz_df[new_column_order]
ppp_yelp_allbiz_df
```

```python
ppp_yelp_allbiz_df = ppp_yelp_allbiz_df.drop_duplicates(
  subset = ['Address', 'Business Name (PPP)'],
  keep = 'first').reset_index(drop = True)
ppp_yelp_allbiz_df
```

```python
count_open1 = ppp_yelp_allbiz_df[ppp_yelp_allbiz_df["Business Status"] == "Open"]
count_open1
```

```python
output_data_file = "ppp_yelp_left_ppp.csv"
# Export the City_Data into a CSV.
ppp_yelp_left_ppp.to_csv(output_data_file, index=False)
```

```python
output_data_file = "ppp_yelp_allbiz.csv"
# Export the City_Data into a CSV.
ppp_yelp_allbiz_df.to_csv(output_data_file, index=False)
```
