```python
yelp_data_df = pd.read_csv("yelp_all_food.csv")
yelp_data_df.head()
```

```python
all_oregon_restaurant_df = yelp_data_df.sort_values(by=['Business Name.1'])
all_oregon_restaurant_df
```

```python
new_column_order = ["Business Name.1", "Address", "City", "State", "Zip", "Business Status", "Star Rating"]
all_oregon_restaurant_df = all_oregon_restaurant_df[new_column_order]
print("All Oregon Food Establishment Yelp Data")
all_oregon_restaurant_df.head()
```

```python
all_oregon_restaurant_df = all_oregon_restaurant_df[all_oregon_restaurant_df["State"] == "OR"]
print("All Yelp Food Data for Oregon")
all_oregon_restaurant_df
```

```python
Address = []
for index in range(len(all_oregon_restaurant_df)):
    address = f'{all_oregon_restaurant_df.iloc[index, 1]} , {all_oregon_restaurant_df.iloc[index, 2]}, {all_oregon_restaurant_df.iloc[index, 3]} {all_oregon_restaurant_df.iloc[index, 4]} ,  United States'
    Address.append(address)
   
len(Address)
```

```python
all_oregon_restaurant_df['Address'] = Address
new_column_order = ["Business Name.1", "Address", "Business Status", "Star Rating"]
all_oregon_restaurant_df = all_oregon_restaurant_df[new_column_order]
print(all_oregon_restaurant_df["Business Name.1"].count())
all_oregon_restaurant_df.head(20)
```

```python
all_oregon_restaurant_df.drop_duplicates(subset=['Address'], keep='first', inplace=True)
all_oregon_restaurant_df
```

```python
all_oregon_restaurant_df["Address"] = all_oregon_restaurant_df["Address"].str.upper()
all_oregon_restaurant_df
```

```python
all_oregon_restaurant_df = all_oregon_restaurant_df.rename(columns={"Business Name.1":"Business Name"})
all_oregon_restaurant_df
```

```python
output_data_file = "final_oregon_restaurant_yelp_data.csv"
# Export the City_Data into a CSV.
all_oregon_restaurant_df.to_csv(output_data_file, index=False)
```
