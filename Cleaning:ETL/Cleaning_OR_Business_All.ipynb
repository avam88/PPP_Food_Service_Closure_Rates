```python
# Add the Pandas dependency.
import pandas as pd
```

```python
all_oregon_business_df = pd.read_csv("data/Active_Businesses_-_ALL.csv")
all_oregon_business_df.head()
```

```python
new_column_order = ["Business Name", "Address", "City", "State", "Zip", "Registry Date", "Associated Name Type"]
all_oregon_business_df = all_oregon_business_df[new_column_order]
print("All Oregon Business Data")
all_oregon_business_df.head()
```

```python
# dropping duplicate business entity entries
all_oregon_business_mailing_df = all_oregon_business_df[all_oregon_business_df["Associated Name Type"] == "MAILING ADDRESS"]
print(all_oregon_business_df["Business Name"].count())
print("All Active Business Data for Oregon")
all_oregon_business_mailing_df
```

```python
Address = []
for index in range(len(all_oregon_business_mailing_df)):
    address = f'{all_oregon_business_mailing_df.iloc[index, 1]} , {all_oregon_business_mailing_df.iloc[index, 2]}, {all_oregon_business_mailing_df.iloc[index, 3]} {all_oregon_business_mailing_df.iloc[index, 4]} ,  United States'
    Address.append(address)
   
len(Address)
```

```python
all_oregon_business_mailing_df['Address'] = Address
new_column_order = ["Business Name", "Address", "Registry Date"]
all_oregon_business_mailing_df = all_oregon_business_mailing_df[new_column_order]
all_oregon_business_mailing_df["Address"] = all_oregon_business_mailing_df["Address"].str.upper()
print(all_oregon_business_mailing_df["Business Name"].count())
all_oregon_business_mailing_df.head(20)
```

```python
output_data_file = "final_oregon_allbiz_mailling_data.csv"
# Export the City_Data into a CSV.
all_oregon_business_mailing_df.to_csv(output_data_file, index=False)
```
