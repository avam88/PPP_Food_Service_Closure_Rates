```python
import pandas as pd
from pathlib import Path
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from datetime import datetime


from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
```

```python
ppp_df = pd.read_csv(Path('data/ppp_yelp_allbiz.csv'))
ppp_df.head()
```

```python
# code Business Satus to binary value
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
ppp_copy = ppp_df.copy()
ppp_copy['Business Status'] = le.fit_transform(ppp_copy['Business Status'])
ppp_copy
# 0 = Closed, 1 = Open
```

```python
# convert Approval Date to datetime format
ppp_copy["Approval Date"] = pd.to_datetime(ppp_copy["Approval Date"])
print(ppp_copy.dtypes)
ppp_copy
```

```python
# convert Approval Date datetime to ordinal format
ppp_copy["Approval Date"] = ppp_copy["Approval Date"].apply(lambda x: x.toordinal())
ppp_copy
```

```python
# establish contextual "ordinal" origin date of 2020-01-01

import datetime
 
# using date.today() to get todays date
origin_date = pd.to_datetime('2020-01-01')
 
# Using toordinal() to generate ordinal value.
toOrdinal = origin_date.toordinal()
 
# Prints Ordinal Value of Todays Date.
print(f"Ordinal of date {origin_date} is {toOrdinal}")
```

```python
# establishing loan approval date as a count of days past January 1st, 2020
ppp_copy["Approval Date"] = ppp_copy["Approval Date"] - 737425
ppp_copy
```

```python
new_column_order = ["Loan Amount", "Foregiveness Amount", "Foregiveness Rate", "Approval Date", "Business Status", "Jobs"]
ppp_copy = ppp_copy[new_column_order]
ppp_copy
```

```python
# Define the features set.
X = ppp_copy.drop(columns = ["Business Status"])
y = ppp_copy["Business Status"]
print(X.shape)
print(y.shape)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=78, train_size=0.80)
print(X_train.shape)
print(y_train.shape)
```

```python
# Creating a StandardScaler instance.
scaler = StandardScaler()
# Fitting the Standard Scaler with the training data.
X_scaler = scaler.fit(X_train)

# Scaling the data.
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
# Create a random forest classifier.
rf_model = RandomForestClassifier(n_estimators=128, random_state=78) 
```

```python
# Fitting the model
rf_model = rf_model.fit(X_train_scaled, y_train)
```

```python
# Making predictions using the testing data.
predictions = rf_model.predict(X_test_scaled)
```

```python
# Calculating the confusion matrix.
cm = confusion_matrix(y_test, predictions)

# Create a DataFrame from the confusion matrix.
cm_df = pd.DataFrame(
    cm, index=["Actual 0", "Actual 1"], columns=["Predicted 0", "Predicted 1"])

cm_df
```

```python
# Calculating the accuracy score.
acc_score = accuracy_score(y_test, predictions)
```

```python
# Displaying results
print("Confusion Matrix")
display(cm_df)
print(f"Accuracy Score : {acc_score}")
print("Classification Report")
print(classification_report(y_test, predictions))
```

```python
# Calculate feature importance in the Random Forest model.
importances = rf_model.feature_importances_
importances
```

```python
# We can sort the features by their importance.
sorted(zip(rf_model.feature_importances_, X.columns), reverse=True)
```
